```python
import pandas as pd
import numpy as np
```

### paso 1 carga, limpieza y transformacion de datos 

```python
df_canada = pd.read_excel('Canada.xlsx', sheet_name='Canada by Citizenship',
                       skiprows= range(20), skipfooter=2)
df_canada
```

```python
print(df_canada.shape)
```

```python
df_canada.drop(['AREA','REG','DEV','Type','Coverage'],axis=1,inplace=True)
df_canada.head(5)
```

```python
df_canada.rename(columns = {'OdName':'Country','AreaName':'Continent','RedName':'Region'},inplace = True)
df_canada.head()
```

```python
## verificar los tipos los tipos de columna que sean string 
all(isinstance(column,str)for column in df_canada.columns)
```

```python
df_canada.columns =list(map(str,df_canada.columns))
```

```python

all(isinstance(column,str)for column in df_canada.columns)
```

```python
df_canada.set_index('Country', inplace = True)
df_canada.head(5)
```

```python

df_canada['Total']=df_canada.sum(axis=1)
df_canada.head()
```

```python
years = list(map(str,range(1980,2014)))
years
```

### parte 2 vizualizacion de datos con mapplotlib

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
```

### graficos de area

```python
df_canada.sort_values(['Total'],ascending=False,
                     axis=0,inplace=True)
df_canada.head(10)
```

```python
df_top5=df_canada.head()
df_top5
```

```python
## transponer renglones y columnas 
df_top5=df_top5[years].transpose()
df_top5
```

```python
df_top5.index=df_top5.index.map(int)
df_top5
```

```python
df_top5.plot(kind='area',
            stacked=False,
            figsize=(20,10))
plt.title('Tendencias de inmigracion de los primeros 5 paises')
plt.ylabel('numero de inmigrantes ')
plt.xlabel('años')
plt.show()
```

```python
## tranpariencia
df_top5.plot(kind='area',
             alpha=0.25,
            stacked=False,
            figsize=(20,10))
plt.title('Tendencias de inmigracion de los primeros 5 paises')
plt.ylabel('numero de inmigrantes ')
plt.xlabel('años')
plt.show()
```

```python
ax= df_top5.plot(kind = 'area',
                alpha=0.35,
                figsize=(20,10))
ax.set_title('Tendencias de inmigracios del top 5 de paises a canada ')
ax.set_ylabel('numero de inmigrantes ')
ax.set_xlabel('años')
```

```python
df_chile = df_canada.loc['Chile',years]
df_chile.head()
```

```python
df_chile.plot(kind='bar',figsize=(10,6))
plt.xlabel('Años')
plt.ylabel('Numero de inmigrnates ')
plt.title('inmigrantes de 1980 al 2013')
plt.show()
```

```python
df_Mexico = df_canada.loc['Mexico',years]
df_Mexico.head()
```

```python
df_Mexico.plot(kind='barh',figsize=(10,6))
plt.ylabel('Años ')
plt.xlabel('Numero de inmigrnates')

plt.title('inmigrantes de 1980 al 2013')
plt.show()
```

### graficos circulares 


```python
df_continentes= df_canada.groupby('Continent',axis=0).sum()
df_continentes
```

```python
df_continentes['Total'].plot(kind='pie',
                            figsize=(5,6),
                            autopct='%1.1f%%',
                            startangle=0,
                            shadow=True)
plt.title('inmigracion a canada por continente 1980-2013')
plt.show()
```

```python
explode_list=[0.1,0,0,0,0.1,0.1]
df_continentes['Total'].plot(kind='pie',
                            figsize=(15,6),
                            autopct='%1.1f%%',
                            startangle=0,
                            shadow=True,
                            labels=None,
                            pctdistance=1.15,
                            explode= explode_list)
plt.title('inmigracion a canada por continente 1980-2013')
plt.axis('equal')
plt.legend(labels=df_continentes.index, loc='upeer right')
plt.show()
```
